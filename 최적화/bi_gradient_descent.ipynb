<b><h3> Multivariate(Bi) Gradient Descent</h3></b>

Bi Gradient Descent는 2개의 변수에 대한 Gradient Descent를 의미
&nbsp;

최적점은 fp(x) = 0일 때 발생
&nbsp;

gradient descent는 x와 fp(x)의 차이가 얼마나지 않을 때 <b>fp는 0에 가깝다는 것을 의미</b>하고 최적화된 값이라는 의미
&nbsp;

h: step size 
&nbsp;

tol: x와 oldx의 차이로 일정 tol보다 작으면 x와 oldx값의 차이가 얼마 나질 않아서 값은 <b>수렴</b>한다는 의미

```r
bi_gradient_descent = function(fp, x, h=1e-2, tol=1e-4, m=1e3 ){
  iter=0
  oldx = x
  x = x-h*fp(x) # initial gradient만큼 이동
  vecnorm = sqrt(sum((x-oldx)^2))

  while(vecnorm > tol){
    iter = iter+1
    if(iter>m) stop('max iteration')
    oldx = x
    x = x-h*fp(x)
    cat('At iteration ', iter, ' value of x1: ',x[1], ', x2: ',x[2], '\n',sep='')
    venorm = sqrt(sum((x-oldx)^2))
  }
  return(x)
}
```

```r
fp = function(x){
  x1 = 2*x[1]^2+1
  x2 = x[2]^3
  return(c(x1, x2))
}
```

```r
bi_gradient_descent(fp, x=c(4,1))
```
