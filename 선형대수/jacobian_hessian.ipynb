<b><h3> Jacobian & Hessian matrix</h3></b>

Jacobian matrix: 1차 미분행렬
&nbsp;

Hessian matrix: 2차 미분행렬

```r
library(numDeriv)
```

```r
func = function(x){
  a = x[1]
  b = x[2]
  rez = (a^2)*b^3
}
```

```r
# 1차 미분 행렬
jacobian = grad(func, c(1,2))
jacobian
```

```r
# 2차 미분 행렬
hessian = hessian(func, c(1,2))
hessian
```

<b><h4> image data에 Jacobian & Hessian matrix 적용</h4></b>

```r
# install.packages('imager')
library(imager)
```

```r
wd = 'D:\\R\\image\\kpu.jpg'
im = load.image(wd)
plot(im, main='KPU')
```

```r
# jacobian matrix
grad = imgradient(im)
plot(grad)
```

```r
# hessian matrix
hessian = imhessian(im)
plot(hessian)
```
