<b><h3> Genetic Algorithm </h3></b>

```r
library(GA)
```

```r
# rastrigin function
rastrigin = function(x1,x2){20+x1^2 + x2^2 -10*(cos(2*pi*x1)+cos(2*pi*x2))}
x1 = seq(-5,5,by = 0.1)
x2 = seq(-5,5,by = 0.1)

f = outer(x1,x2, rastrigin)
```

```r
# rastrigin 시각화
persp3D(x1,x2,f, theta = 50, phi=20)
```

```r
# GA
ga = ga(type='real-valued', fitness= function(x) -rastrigin(x[1], x[2]), lower=c(-5,-5), upper=c(5,5), popSize=1000, maxiter=100, pcrossover=.8, pmutation=.1)
```

```r
# 함수 수렴 시각화
plot(ga)
```

```r
# GA 함수 파라미터
summary(ga)
```

```r
# 최적해
ga@solution
```

```r
# rastrigin 적용
rastrigin(ga@solution[1], ga@solution[2])
```

```r
# 지수표기를 숫자표기로
options('scipen' = 100)
str(rastrigin(ga@solution[1], ga@solution[2]))
```
